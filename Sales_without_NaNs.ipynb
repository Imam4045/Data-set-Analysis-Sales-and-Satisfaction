## **Range of Columns that have Numerical Values** ##

```python
import pandas as pd

df = pd.read_csv('Sales_without_NaNs_v1.3.csv')

print("--------Sales Before--------")

sales_bf_min = df['Sales_Before'].min()
print(f'Minimum Range: ', round(sales_bf_min,2))

sales_bf_mx = df['Sales_Before'].max()
print(f'Maximum Range: ', round(sales_bf_mx,2))
print(end='\n')

print("--------Sales After--------")

sales_after_min = df['Sales_After'].min()
print(f'Minimum Range: ', round(sales_after_min,2))

sales_after_mx = df['Sales_After'].max()
print(f'Maximum Range: ', round(sales_after_mx,2))
print(end='\n')

print("--------Customer Satisfaction Before--------")

customer_st_bf_min = df['Customer_Satisfaction_Before'].min()
print(f'Minimum Range: ', round(customer_st_bf_min,2))

customer_st_bf_mx= df['Customer_Satisfaction_Before'].max()
print(f'Maximum Range: ', round(customer_st_bf_mx,2))
print(end='\n')

print("--------Customer Satisfaction After--------")

customer_st_af_min = df['Customer_Satisfaction_After'].min()
print(f'Minimum Range: ', round(customer_st_af_min,2))

customer_st_af_mx= df['Customer_Satisfaction_After'].max()
print(f'Maximum Range: ', round(customer_st_af_mx,2))
print(end='\n')
```

## **Drawing Line Graph** ##

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV file
file_path = "Sales_without_NaNs_v1.3.csv"
data = pd.read_csv(file_path)

# Select required columns
stats_data = data[["Sales_Before", "Sales_After",
                   "Customer_Satisfaction_Before", "Customer_Satisfaction_After"]]

# Calculate statistics
stats_summary = pd.DataFrame({
    "Mean": stats_data.mean(),
    "Median": stats_data.median(),
    "Variance": stats_data.var(),
    "Standard Deviation": stats_data.std()
})

# Transpose for plotting
stats_summary = stats_summary.T

# Plot
plt.figure(figsize=(12, 8))
for column in stats_summary.columns:
    plt.plot(stats_summary.index, stats_summary[column], marker="o", label=column)

plt.title("Statistics for Numerical Columns")
plt.xlabel("Statistic Type")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

## **Label Encoding** ##

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Sample data
data = pd.DataFrame({'Fruit': ['Apple', 'Banana', 'Mango', 'Apple', 'Banana']})

# Apply label encoding
le = LabelEncoder()
data['Fruit_encoded'] = le.fit_transform(data['Fruit'])

print(data)
```

## **One-Hot Encoding** ##

```python
import pandas as pd

# Same fruit data
data = pd.DataFrame({'Fruit': ['Apple', 'Banana', 'Mango', 'Apple', 'Banana']})

# Apply one-hot encoding with pandas and force integer type
one_hot = pd.get_dummies(data['Fruit'], prefix='Fruit', dtype=int)
data = pd.concat([data, one_hot], axis=1)

print(data)
```

## **Ordinal Encoding** ##

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# Sample car data
data = pd.DataFrame({'Fuel_Efficiency': ['Medium', 'High', 'Low', 'Medium', 'High']})

# Define order
efficiency_order = ['Low', 'Medium', 'High']

# Apply ordinal encoding
encoder = OrdinalEncoder(categories=[efficiency_order])
data['Fuel_Efficiency_Encoded'] = encoder.fit_transform(data[['Fuel_Efficiency']])

print(data)
```

## **Frequency Encoding** ##

```python
import pandas as pd

# Sample data
data = pd.DataFrame({'Car_Brand': ['Toyota', 'Honda', 'Ford', 'Toyota', 'Ford', 'Honda', 'Honda']})

# Count/frequency encoding
freq_encoding = data['Car_Brand'].value_counts().to_dict()
data['Car_Brand_Freq'] = data['Car_Brand'].map(freq_encoding)

print(data)
```
